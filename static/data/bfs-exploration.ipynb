```python
import pandas as pd

df = pd.read_csv('hero-network.csv')
```

```python
df
```

```python
hero_circles = {}
for i in df.index:
    if df['hero1'][i] not in hero_circles:
        hero_circles[df['hero1'][i]] = set()
    hero_circles[df['hero1'][i]].add(df['hero2'][i])
    if df['hero2'][i] not in hero_circles:
        hero_circles[df['hero2'][i]] = set()
    hero_circles[df['hero2'][i]].add(df['hero1'][i])
```

```python
hero_circles ['SPIDER-MAN/PETER PAR']
```

```python
smallest_circle = 1000
smallest_circle_hero = ''
for hero in hero_circles:
    circle_size = len(hero_circles[hero])
    if circle_size < smallest_circle:
        smallest_circle = circle_size
        smallest_circle_hero = hero
```

```python
def bfs(start_hero, end_hero):
    visited = []
    queue = [[start_hero]]
    if start_hero == end_hero:
        return(None)
    while queue:
        path = queue.pop(0)
        hero = path[-1]
        if hero not in visited:
            neighbors = hero_circles[hero]
            for neighbor in neighbors:
                new_path = list(path)
                new_path.append(neighbor)
                queue.append(new_path)
                if neighbor == end_hero:
                    return(new_path)
            visited.append(hero)

    return(None)


bfs('REAGAN, RONALD', 'HITLER, ADOLF')
```

```python
def furthest_bfs(start_hero):
    longest_path = 100
    longest_path_hero = ''
    for hero in hero_circles:
        path = bfs(start_hero, hero)
        if path == None:
            continue
        path_length = len(path)
        
        if path_length > longest_path:
            longest_path = path_length
            longest_path_hero = hero
        print(path)
    print('done')
    return longest_path_hero, longest_path

furthest_bfs('LUDLUM, ROSS')
```
